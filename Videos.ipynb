```python
from collections import OrderedDict
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

from arrow import now
def date():return now('America/New_York').format('YYYY-MM-DD')
def time():return now('America/New_York').format('HH:mm')

import pandas as pd
from pandas import DataFrame
```

```python
HEADLESS = True
if HEADLESS:
    #https://stackabuse.com/getting-started-with-selenium-and-python/
    from selenium.webdriver.chrome.options import Options
    options = Options()
    options.headless = True
    driver = webdriver.Chrome(options=options)
    driver.set_window_size(1440, 900)
else:
    driver=webdriver.Chrome()
    
print('Headless', HEADLESS)
```

```python
URL='https://www.youtube.com/channel/UCKdiWmb6xNpK4d7MANT3Ghw/about'

driver.get(URL)
subscribers = driver.find_element_by_id('subscriber-count').text
subscribers = round(float(subscribers.split()[0].replace('K',''))*1000)
views =  driver.find_element_by_id('right-column').text.split()[-2]
established = driver.find_element_by_id('right-column').text.split('\n')[1]
views = int(views.replace(',',''))
channelInfo = OrderedDict(title='e-NABLE',  href=URL, age= established, subscribers = subscribers, views = views, date=date(), time=time())
videos = [channelInfo]

driver.get('https://www.youtube.com/channel/UCKdiWmb6xNpK4d7MANT3Ghw/videos?view=0&sort=dd&flow=grid')
body = driver.find_element_by_tag_name('body')
for i in range(5): 
    body.send_keys(Keys.PAGE_DOWN)
    sleep(0.3)
```

```python
details = body.find_elements_by_id('details')   
videotitles = body.find_elements_by_id('video-title')
for i,n in enumerate(details):
    title, views, age = n.text.split('\n') 
    href = videotitles[i].get_attribute('href')
    videos.append( OrderedDict(title=title, href=href, views=views, age=age))
```

```python
DataFrame(videos).to_excel('videos.xls')
DataFrame(videos)
```
