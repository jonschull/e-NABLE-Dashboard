```python
def getMembers(): 
    i=1
    members = []
    d=date()
    while i>=0:
        driver.get(f'https://hub.e-nable.org/directory/members?page={i}&per-page=25&')
        sleep(2)
        i+=1
        if 'Oooops' in driver.find_element_by_class_name('panel-heading').text:
            i=-1
        else:                        #members are found in all the non-null body elements
            bodies = [body for body in driver.find_elements_by_class_name('media-body') if body.text] 
            for body in bodies:
                name =    body.find_element_by_tag_name('a').text
                groups =  body.find_element_by_tag_name('small').text.split(', ')
                href=     body.find_element_by_tag_name('a').get_attribute('href') 
                member = OrderedDict(date=d, href=href, name=name, groups=groups)
                members.append(member)
                
    return members

members = getMembers()
```

```python
from html import unescape
for i in range(len(members)):
    member=members[i]
    driver.get(member['href'])
    sleep(1)
    stats = driver.find_element_by_id('layout-content').find_element_by_class_name('statistics').text.split()
    member['followerCount'] = int(stats[0])
    member['followingCount'] = int(stats[2])
    member['spacesCount'] = int(stats[4])
    member['spaces'] = [unescape(img.get_attribute('alt')) for img in driver.find_element_by_id('user-spaces-panel').find_elements_by_tag_name('img')]
    #print(member['spaces'])
    members[i] = member
    print(i,end=' ')
```
